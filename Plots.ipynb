```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Data Load

```python
data = {'Evaluated System': ['Local Python', 'Local PySpark', 'Spark w/ 2 Node', 'Spark w/ 4 Node', 'Spark w/ 8 Node'],
        'Time(seconds)': [47.54, 4.36, 9.01, 4.35, 2.92] }

data_load = pd.DataFrame (data, columns = ['Evaluated System','Time(seconds)'])

sns.set(font_scale=1.2, style='white')
ax = sns.barplot(x="Time(seconds)", y="Evaluated System", data=data_load)
ax.set_title('Ingesting Data')
plt.savefig('output/ingestData.png')
plt.show()
```

### Data Imputation

```python
data = {'Evaluated System': ['Local Python', 'Local PySpark', 'Spark w/ 2 Node', 'Spark w/ 4 Node', 'Spark w/ 8 Node'],
        'Time(seconds)': [0.03, 1.41 ,1.33 ,1.32 ,1.26] }

data_impute = pd.DataFrame (data, columns = ['Evaluated System','Time(seconds)'])

sns.set(font_scale=1.2, style='white')
ax = sns.barplot(x="Time(seconds)", y="Evaluated System", data=data_impute)
ax.set_title('Data Imputation')
plt.savefig('output/dataImputation.png')
plt.show()
```

### Local Machine Python

```python
df = pd.DataFrame({
    'Model' : ['Random Forest', 'Gradient Boost'],
    'Training': [867.84, 303.8],
    'Grid Search': [1630.74, 4358.46],
    'Random Search': [626.41, 1111.9]
})
new_df = pd.melt(df, id_vars="Model", var_name="Task", value_name="Time(seconds)")

sns.set(font_scale=1.2, style='white')
g = sns.catplot(data=new_df, kind='bar', x='Task', y="Time(seconds)", hue='Model')
g.set_axis_labels("", "Time (seconds)")
g.fig.suptitle("Python on Local", fontsize=18)

for axes in g.axes.flat:
    _ = axes.set_xticklabels(axes.get_xticklabels(), rotation=45)
    
plt.savefig('output/pythonLocal.png')
plt.show()
```

### PySpark Local

```python
df = pd.DataFrame({
    'Model' : ['Random Forest', 'Gradient Boost'],
    'Training': [4.07, 6.97],
    'Cross Validation': [304.04, 753.39],
    'Train Validation': [66.05, 211.63]
})
new_df = pd.melt(df, id_vars="Model", var_name="Task", value_name="Time(seconds)")

sns.set(font_scale=1.2, style='white')
g = sns.catplot(data=new_df, kind='bar', x='Task', y="Time(seconds)", hue='Model')
g.set_axis_labels("", "Time (seconds)")
g.fig.suptitle("PySpark on Local", fontsize=18)

for axes in g.axes.flat:
    _ = axes.set_xticklabels(axes.get_xticklabels(), rotation=45)

plt.savefig('output/pysparkLocal.png')
plt.show()
```

### PySpark Training: 2 vs 4 vs 8

```python
df = pd.DataFrame({
    'Model' : ['Random Forest', 'Gradient Boost'],
    '2 Nodes': [4.43, 11.37],
    '4 Nodes': [4.56, 11.6],
    '8 Nodes': [4.15, 11.6]
})
new_df = pd.melt(df, id_vars="Model", var_name="Nodes", value_name="Time(seconds)")

sns.set(font_scale=1.2, style='white')
g = sns.catplot(data=new_df, kind='bar', x='Nodes', y="Time(seconds)", hue='Model')
g.set_axis_labels("", "Time (seconds)")
ax = g.facet_axis(0,0)
for p in ax.patches:
    ax.text(p.get_x() - 0.01, 
            p.get_height() * 1.02, 
           '{0:.1f}'.format(p.get_height()),
            color='black', 
            rotation='horizontal', 
            size='medium')
plt.figtext(0.4, 0.01, "Model Training on Databricks", ha="center", 
            fontsize=18, bbox={"facecolor":"limegreen", "alpha":0.5, "pad":5})
plt.show()
```
